# Expression Maximizer

## üìù Project Description

In the internal files of **Shaparak (Electronic Payment Network Company)**, a mathematical expression was found containing missing operators and no parentheses. Due to system updates, the original arithmetic operators were replaced with the unknown symbol `?`.

You are given an expression in the form:

```
a?b?c
```

where `a`, `b`, and `c` represent natural numbers.
The goal is to replace each `?` with either `+` or `√ó`, then apply parentheses in the best possible way to achieve **the maximum final value** of the expression.

Your task is to write a Python program that calculates the highest obtainable result.


### üîπ Input

A single line containing an expression in the form:

```
a?b?c
```

where `a`, `b`, and `c` are natural numbers.

### üîπ Output

Print one integer ‚Äî the maximum achievable value after replacing `?` with `+` or `√ó` and parenthesizing optimally.



### ‚ú® Example 1

**Input:**

```
5?3?2
```

**Output:**

```
30
```

Explanation:
Replacing both `?` with `√ó` results in the expression:

```
5 √ó 3 √ó 2 = 30
```

which is the maximum possible value.


### ‚ú® Example 2

**Input:**

```
14?100?1
```

**Output:**

```
1414
```

Explanation:
The maximum result is achieved by choosing:

```
14 √ó (100 + 1) = 1414
```

Here, the first `?` becomes `√ó` and the second `?` becomes `+`, with optimal parentheses applied.

# Solution

````python
# Read input
string_with_qs = input()
print(string_with_qs)
````

````python
# Extract numbers from the input string
a, b, c = map(int, string_with_qs.split('?'))
````

````python
print(
    max(
        [
            (a * b * c),
            a * (b + c),
            (a + b) * c,
            a + b + c,
        ]
    )
)
````
