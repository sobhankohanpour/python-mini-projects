# Binary String Classifier

## üìù Project Description

You are given a dataset containing `n` binary strings, each of length `‚Ñì`. Each string is labeled with either `Y` or `N`.

Now, you are given `q` binary strings, also of length `‚Ñì`. Your task is to determine for each query string whether it exists in the dataset and print its corresponding label. If a query string does not exist in the dataset, output `"Unknown"`.



### üîπ Input

1. The first line contains three integers separated by spaces: `n`, `q`, and `‚Ñì`

   ```
   1 ‚â§ n, q ‚â§ 10,000
   1 ‚â§ ‚Ñì ‚â§ 100
   ```
2. The next `n` lines each contain a binary string of length `‚Ñì` followed by a space and its label (`Y` or `N`). All `n` strings are unique.
3. The following `q` lines each contain a binary string of length `‚Ñì` for which you need to determine the label.


### üîπ Output

Print `q` lines. Each line should contain:

* The label (`Y` or `N`) if the query string exists in the dataset.
* `"Unknown"` if the string does not exist in the dataset.

### üîπ Example 1

**Input:**

```
5 4 3
000 Y
011 N
100 Y
010 Y
110 N
011
110
111
100
```

**Output:**

```
N
N
Unknown
Y
```

**Explanation:**

* `"011"` exists in the dataset with label `N`.
* `"110"` exists with label `N`.
* `"111"` does not exist ‚Üí `"Unknown"`.
* `"100"` exists with label `Y`.

# Solution

````python
# Read initial inputs
num_entries, num_queries, string_length = map(int, input().split())
````

````python
# Read dataset
dataset = {}
for _ in range(num_entries):
    binary_string, label = input().split()
    dataset[binary_string] = label
````

````python
# Read queries
queries = [input() for _ in range(num_queries)]

# Process queries
for query in queries:
    if query in dataset:
        print(dataset[query])
    else:
        print("Unknown")
````
