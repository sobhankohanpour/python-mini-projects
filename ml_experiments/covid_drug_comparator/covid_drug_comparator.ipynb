# COVID Drug Effectiveness Comparator

## Project Description

Two countries, **Shekarestan** and **Namakestan**, which have roughly equal populations, have used different medications to treat their COVID-19 patients. Both countries have reported the number of confirmed cases and deaths.

The goal of this project is to determine which country’s treatment was more effective. The country with a higher number of recovered patients is considered to have used a more effective treatment. The number of recovered patients in a country is calculated as the difference between the total number of confirmed cases and the number of deaths.


### Input

The input consists of four lines:

1. The first two lines contain two integers, `n` and `k`, representing the number of confirmed cases and deaths in **Shekarestan**, respectively.
   **Constraints:** `1 ≤ k ≤ n ≤ 1000`

2. The third and fourth lines contain two integers, `p` and `q`, representing the number of confirmed cases and deaths in **Namakestan**, respectively.
   **Constraints:** `1 ≤ q ≤ p ≤ 1000`

### Output

Output a single line:

* `"Shekarestan"` if Shekarestan’s treatment was more effective.
* `"Namakestan"` if Namakestan’s treatment was more effective.
* `"Equal"` if both countries have the same number of recovered patients.

### Sample Input 1

```
2
1
4
1
```

### Sample Output 1

```
Namakestan
```

### Sample Input 2

```
3
1
4
2
```

### Sample Output 2

```
Equal
```

### Sample Input 3

```
5
1
4
2
```

### Sample Output 3

```
Shekarestan
```


# Solution

````python
# Input for Shekarestan
sicks_in_shekarestan = int(input())
deads_in_shekarestan = int(input())

print(f"Number of sicks in Shekarestan: {sicks_in_shekarestan}")
print(f"Number of deads in Shekarestan: {deads_in_shekarestan}")

# Input for Namakestan
sicks_in_namakestan = int(input())
deads_in_namakestan = int(input())

print(f"Number of sicks in Namakestan: {sicks_in_namakestan}")
print(f"Number of deads in Namakestan: {deads_in_namakestan}")

# Calculate recoveries
Shekarestan = sicks_in_shekarestan - deads_in_shekarestan
Namakestan = sicks_in_namakestan - deads_in_namakestan

# Compare recoveries and print result
if Shekarestan > Namakestan:
    print("Shekarestan")
elif Shekarestan < Namakestan:
    print("Namakestan")
else:
    print("Equal")
````
